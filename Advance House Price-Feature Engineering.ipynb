# Missing values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in housing.columns if housing[feature].isnull().sum()>1 and housing[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(housing[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(housing,features_nan):
    data=housing.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

housing=replace_cat_feature(housing,features_nan)

housing [features_nan].isnull().sum()
```

```python
housing.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in housing.columns if housing[feature].isnull().sum()>1 and housing[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(housing[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=housing[feature].median()
  
    
    ## create a new feature to capture nan values
    housing[feature+'nan']=np.where(housing[feature].isnull(),1,0)
    housing[feature].fillna(median_value,inplace=True)
    
housing[numerical_with_nan].isnull().sum()
```

```python
housing.head(10)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    housing[feature]=housing['YrSold']-housing[feature]
```

# Numerical Variables

Since the numerical variables are skewed we will perform log normal distribution

```python
housing.head()
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    housing[feature]=np.log(housing[feature])
```

# Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in housing.columns if housing[feature].dtype=='O']
```

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=housing.groupby(feature)['SalePrice'].count()/len(housing)
    temp_df=temp[temp>0.01].index
    housing[feature]=np.where(housing[feature].isin(temp_df),housing[feature],'Rare_var')
```

```python
housing.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=housing.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    housing[feature]=housing[feature].map(labels_ordered)
```

# feature_scaling

```python
feature_scale=[feature for feature in housing.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(housing[feature_scale])
```

```python
scaler.transform(housing[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([housing[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(housing[feature_scale]), columns=feature_scale)],
                    axis=1)
```
